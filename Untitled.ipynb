```python
# Method to extract 'float' from 'object' 

import re

def get_number(name):
    title_search = re.search('([\d+\.+\d]+\W)', name)
    
    if title_search:
        return title_search.group(1)
    return ""
```

```python
import xgboost
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_squared_log_error
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

y = np.log1p(train_data.Price)  #log se negative aa rha hai 
X = train_data.drop(['Price'],axis=1)
# df_test = df_test.drop('New_Price',axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size=0.82,test_size=0.18,random_state=0)
```

```python
y
```

```python
train_data.head()
```

```python
negative_values = y_valid[y_valid < 0]

if len(negative_values) > 0:
    print("Negative values in y_valid:")
    for value in negative_values:
        print(value)
else:
    print("No negative values found in y_valid.")

```

```python
# Define and train the XGBoost model
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)

# Make predictions
predictions = my_model.predict(X_valid)

# Compute evaluation metrics
mae = mean_absolute_error(predictions, y_valid)
mse = mean_squared_error(predictions, y_valid)
msle = mean_squared_log_error(predictions + 1e-10, y_valid + 1e-10)  # Adding a small value to avoid log(0)

# Print evaluation metrics
print("MAE: " + str(mae))
print("MSE: " + str(mse))
print("MSLE: " + str(msle))
```
